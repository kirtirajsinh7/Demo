```python
# 15.Write a Python program to find the greatest common divisor (GCD) of two numbers.
a=12
factor_a=set()
b=18
factor_b=set()
for i in range(1,a+1):
    if a%i==0:
        factor_a.add(i)
for j in range(1,b+1):
    if b%j==0:
        factor_b.add(j)
print(max(factor_a.intersection(factor_b)))
```

```python
# 16.Write a Python program to find the least common multiple (LCM) of two numbers.

# Option: 1
a=4
b=6
lcm=0
for i in range(max(a,b),10000000):
    if i%a==0 and i%b==0:
        lcm=i
        break
print(f"LCM({a},{b})={lcm}")
# Output: LCM(4,6)=12

# Option: 2  # Here We use formula LCM(a,b)= a * b / GCD(a,b)
a=4
factor_a=set()
b=6
factor_b=set()
for i in range(1,a+1):
    if a%i==0:
        factor_a.add(i)
for j in range(1,b+1):
    if b%j==0:
        factor_b.add(j)
GCD=max(factor_a.intersection(factor_b))
lcm=(a*b)//GCD  # We use floor divisior
print(f"LCM({a},{b})={lcm}")

# Option: 3
a=4
b=6

lcm=max(a,b)

while True:
    if lcm%a==0 and lcm%b==0:
        break
    lcm+=1
print(f"LCM({a},{b})={lcm}")

# Conclusion = a * b = LCM(a,b) * GCD(a,b)
```
